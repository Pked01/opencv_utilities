```python
import cv2,time
import IPython.display as Disp
```

```python
vid_name = 'ch01_20190208180212.mp4'

cap = cv2.VideoCapture(vid_name)

def callback():
    print( "clicked!")

def get_time(t1):
    return time.strftime('%H:%M:%S', time.gmtime(t1))
    

!mkdir cut_vids

num_frames = int(cap.get(7))
fps = int(cap.get(5))

cv2.namedWindow('preview',cv2.WINDOW_NORMAL)
# create trackbars for color change
def do_nothing(x):
    pass

cv2.createTrackbar('scroll','preview',0,num_frames,do_nothing)
# cv2.createTrackbar('start_frame','preview',0,num_frames,do_nothing)
# cv2.createTrackbar('end_frame','preview',0,num_frames,do_nothing)
```

```python
### use track bar to slide in video
### press s to start cut of vide
### press e to en the cut
### press c to confirm
vid_idx=0
font = cv2.FONT_HERSHEY_SIMPLEX
while True:
    Disp.clear_output(wait=True)


    ret,frame = cap.read()
    cv2.putText(frame,"1. press s to start ",(20,20), font, 1,(0,0,255),2,cv2.LINE_AA)
    cv2.putText(frame,"2. press e to start ",(20,40), font, 1,(0,0,255),2,cv2.LINE_AA)
    cv2.putText(frame,"3. press c to confirm ",(20,60), font, 1,(0,0,255),2,cv2.LINE_AA)

    ### current frame Position
    c_f = int(cv2.getTrackbarPos('scroll','preview'))
    cap.set(1,c_f)
    
    cv2.imshow('preview',frame)
    k = cv2.waitKey(1)
    if ((k==83) or (k==115)):#press s
        t1 = c_f
    if ((k==69) or (k==101)):#press e
        t2 = c_f    
    if ((k==67)) or ((k==99)):# press c
        vid_idx+=1
        start_time  =  get_time(t1/fps)
        num_second = (t2-t1)/fps
        file_name = 'cut_vids/out_'+str(vid_idx)+'.mp4'
        !ffmpeg -ss {start_time} -i {vid_name} -t {num_second} -c:v libx264 -c:a aac -strict experimental -b:a 128k {file_name} -y 
    if k==27:
        cv2.destroyAllWindows()
        cap.release()
        break
        
        
    
    
```
